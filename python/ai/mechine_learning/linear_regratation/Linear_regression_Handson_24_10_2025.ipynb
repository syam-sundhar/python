What is VIF?(Variance Inflation factor):

Multicolinearity occurs when two or more  independent columns in a dataset are highly correlated with each other.

VIF isused to detect the multicolinear columns.

If VIF score<6:It means that columns are good to go.

If VIF Score>6:It means that columns are Multicolinear columns.

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
#create a list of all numerical columns
col_list=[]
for i in df.columns:
  if ((df[i].dtype!='object') & (i!='charges')):
    col_list.append(i)

#Apply vif on columns present inside the col_list
X=df[col_list]  #Store all trhe columns in the variable x
vif_data=pd.DataFrame()
vif_data['Column_Name']=X.columns
vif_data['VIF_Score']=[variance_inflation_factor(X.values,i) for i in range(len(X.columns))]
vif_data
```

# **Model Building**

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()   #Give a name to my model
```

```python
#Train the model
model.fit(x_train,y_train)
```

```python
#Test the model
y_pred=model.predict(x_test)
```

```python
#now compare it
new_df=pd.DataFrame()
new_df['Actual Charges']=y_test
new_df['Predicted Charges']=y_pred
new_df
```
