# **VIF**

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
col_list=[]
for i in df.columns:
  if (i!='Price'):
   col_list.append(i)
#Apply vif
X=df[col_list]
vif_data=pd.DataFrame()
vif_data['Column_name']=X.columns
vif_data['VIF_score']=[variance_inflation_factor(X.values,i) for i in range(len(X.columns))]
vif_data
```

# **Build Model**

```python
x=df[['Brand','EngineSize','Horsepower','FuelType','Transmission','OwnerCount']]
y=df['Price']
```

```python
#Split the data into training and testing
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
le=LinearRegression()
```

```python
le.fit(x_train,y_train)
```

```python
y_pred=le.predict(x_test)
```

```python
y_pred
```

```python
y_test
```

```python
df1=pd.DataFrame()
df1['Actual_Price']=y_test
df1['Predicted_Price']=y_pred
df1
```

```python
from sklearn.metrics import r2_score
print(r2_score(y_test,y_pred)*100)
```
