# **Elbow Method:**

A graph between your k value and WCSS

WCSS-Within Cluster Sum of Squared Distance

```python
from sklearn.cluster import KMeans
```

```python
k_values=[2,3,4,5,6,7]
```

```python
#Elbow
ssd=[]
for k in k_values:
  km=KMeans(n_clusters=k,max_iter=150,random_state=32)
  km.fit(df_scaled)
  ssd.append(km.inertia_)
```

```python
plt.plot(k_values,ssd)
```

image.png

```python
from sklearn.metrics import silhouette_score

k_values=range(2,8)

silhoutte_scores=[]
for k in k_values:
  kmeans=KMeans(n_clusters=k,random_state=32)
  kmeans.fit(df_scaled)
  sihoutette_avg=silhouette_score(df_scaled,kmeans.labels_)
  silhoutte_scores.append(sihoutette_avg)
```

```python
plt.subplot(1,2,2)
plt.plot(k_values,silhoutte_scores,marker='o',color='green')
plt.title('Silhoutte Score Vs K')
plt.xlabel('Number of clusters(k)')
plt.ylabel('Silhoutte Score')
plt.show()
```

```python
#let's build the final model
kmeans=KMeans(n_clusters=4,max_iter=150,random_state=32)
```

```python
kmeans.fit(df_scaled)
```

```python
print(kmeans.labels_)#this will assign each data points into 4 different clusters we have created
```

```python
df['ClusterID']=kmeans.labels_
df
```
