# VIF:Variance Inflation Factor

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
col_list=[]
for i in df.columns:
  if ((df[i].dtype !='object')& (i!='y')):
    col_list.append(i)

X=df[col_list]
vif_data=pd.DataFrame()
vif_data['Column_Name']=X.columns
vif_data['VIF_Score']=[variance_inflation_factor(X.values,i) for i in range(len(X.columns))]
```

# RFE-
Recursive Feature Elimination

```python
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
x
```

```python
y
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()
```

```python
from sklearn.feature_selection import RFE
```

```python
rfe=RFE(model)
```

```python
rfe.fit(x,y)
```

```python
rfe.support_
```

```python
selected_indices=[i for i,data in enumerate(rfe.support_) if data==True]
selected_indices
```

```python
column_name=x.columns
```

```python
selected_columns=[column_name[i] for i in selected_indices]
selected_columns
```

```python
x=x.loc[:,['education',
 'default',
 'contact',
 'campaign',
 'previous',
 'poutcome',
 'emp.var.rate']]
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=42)
```

```python
x_train
```

```python
y_train
```

```python
x_test
```

```python
y_test
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
```

```python
lr.fit(x_train,y_train)   #training my model
```

```python
y_pred=lr.predict(x_test)   #testing my model
```

```python
new_df=pd.DataFrame()
new_df['Actual Value']=y_test
new_df['Predicted Value']=y_pred
```

```python
new_df
```

```python
from sklearn.metrics import *
```

```python
confusion_matrix(y_test,y_pred)
```

```python
accuracy_score(y_test,y_pred)
```

```python
precision_score(y_test,y_pred)
```

```python
recall_score(y_test,y_pred)
```
