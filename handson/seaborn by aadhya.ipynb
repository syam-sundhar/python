```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data= sns.load_dataset("iris")
data
```

```python
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
sns.histplot(data=data,x="sepal_length",bins=30,color='blue',kde = True)
plt.show()
```

```python
#histogram
sns.histplot(data=data,x="sepal_length",bins=30,color='blue',kde = True)
plt.show()
```

```python
#barplot
sns.barplot(data=data,x='species',y='sepal_length',hue='species',palette='dark')
plt.show()
```

```python
sns.boxplot(data=data,x='species',y='sepal_length',hue='species',palette='coolwarm')
```

```python
sns.scatterplot(data=data,x='sepal_length',y='petal_length',hue='species',palette='Dark2')
```

```python
sns.violinplot(data=data,x='species',y='sepal_length',hue='species',palette='magma')
```

```python
x= data.select_dtypes(include=['float64','int64'])
correlation=x.corr()
sns.heatmap(correlation,annot=True,cmap='coolwarm',vmin=-1,vmax=1)
plt.show()
```

```python
sns.regplot(data=data,x='sepal_length',y='sepal_width',color='blue')
```

```python
sns.relplot(data=data,x='sepal_length',y='sepal_width',hue='species',kind="line",palette="Set1")
```
