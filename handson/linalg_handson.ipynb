```python
def check_transpose(matrix):
    trans=matrix.T
    print(trans)
    if np.array_equal(matrix,trans):
        print("it is symmetric")
        print("____________________")
    elif np.array_equal(matrix,(-1)*trans):
        print("it is skew symmetric")
        print("____________________")
    else:
        print("it is non-symmertric")
        print("____________________") 
m3=np.array([[1, 2], [2, 1]])
m4= np.array([[0, 2], [-2, 0]])
check_transpose(m3)
check_transpose(m4)
```

```python
A = np.array([[2, 5, 3],[4, 3, 3],[5, 6, 3]])

print(f"Matrix:{A}")
print(f"Determinant: {np.linalg.det(A)}")
print(f"Rank: {np.linalg.matrix_rank(A)}")

try:
    print(f"Inverse:{np.linalg.inv(A)}")
except np.linalg.LinAlgError:
    print("Inverse: Matrix is singular")
```

```python
import numpy as np

# Test matrix
A = np.array([[2, 5, 3], [4, 3, 3], [5, 6, 3]])
w, v = np.linalg.eig(A)

print(f"Matrix A:{A}")
print(f"Eigenvalues (λ):{w}")
print(f"Eigenvectors (v):{v}")
fir_eigepai = np.allclose(A @ v[:, 0], w[0] * v[:, 0])
print(f"Verification of first eigenpair (A @ v ≈ λ * v): {fir_eigepai}")
```
