```python
arr = np.append(arr,6) # append --> by default elements will get added at the end of the array.
arr
```

```python
arr = np.insert(arr,0,10)
arr
```

insert the element 30, before 4.

```python
arr = np.insert(arr,3,30)
arr
```

```python
arr[arr==5]=50
arr
```

```python
arr = np.delete(arr,np.where(arr==50))
print(arr)
```
