```python
res = []
for i in df1['acceleration']: 
    if i%2 == 0:
        res.append('Even')
    else:
        res.append('Odd')
df1['col1'] = res
df1.head()
```

```python
def evn_odd(num):
    if num%2 == 0:
        return 'Even'
    else:
        return 'Odd'
```

```python
evn_odd(12)
```

```python
df1['col2'] = df1['acceleration'].apply(evn_odd)
df1.head()
```

```python
# lambda 
# anonymous function / one time execution 

df1['col3'] =  df1['acceleration'].apply(lambda num: 'Even' if num%2 ==0 else 'Odd')
df1.head()
```

```python
df1['col4'] =  df1['acceleration'].apply(lambda num: num**2)
df1.head(2)
```
