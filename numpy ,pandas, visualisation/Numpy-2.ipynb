#### Array Mathematics

```python
ar1 = np.array([[10,20], [30,40]])
ar1
```

```python
np.sum(ar1)
```

```python
np.sum(ar1, axis = 0)
```

```python
np.sum(ar1, axis = 1)
```

```python
print(np.add(20,10))
print(np.subtract(20,10))
print(np.multiply(20,10))
print(np.divide(20,10))
```

```python
print(np.log(20))
print(np.sqrt(20))
print(np.exp(20))
print(np.sin(20))
```
