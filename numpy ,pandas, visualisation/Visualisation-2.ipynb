#### Scatter Plot

```python
# used between two continuous columns
# to understand relation between two continuous columns
```

```python
# mileage vs weight
# if weight is increasing then mileage will reduce ( negative relation / inversely proportional ) # we can validate it 
```

```python
plt.scatter(df1['mpg'], df1['weight'])
```

```python
plt.scatter(df1['mpg'],df1['displacement'])
plt.xlabel('mpg')
plt.ylabel('Displ')
plt.title('Scatter Plot')
plt.show()
```

```python
plt.subplot(121)
plt.scatter(df1['mpg'], df1['weight'])

plt.subplot(122)
plt.scatter(df1['mpg'],df1['displacement'])
plt.xlabel('mpg')
plt.ylabel('Displ')
plt.title('Scatter Plot')

plt.show()
```

```python
A = [21,22,23,24,25]
B = [24,25,22,24,21]
C = [1,2,3,4,5]

plt.subplot(221)
plt.scatter(C,A, marker = 'D', color = 'Black')

plt.subplot(222)
plt.plot(C,A, linestyle = 'dashed', color = 'Red', marker = '+') # line plot
plt.ylabel('Yaxis')
plt.xlabel('Xaxis')

plt.subplot(223)
plt.scatter(C,B)

plt.subplot(224)
plt.stackplot(C,B) # stack plot

plt.subplots_adjust(wspace = 0.3, hspace = 0.3)
plt.show()
```
