```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df1 = pd.read_csv('Auto.csv')
df1.head()
```

```python
# Divide columns using statistics data types
# continous: mpg, displacement, Horse Power, weight , acceleration
# discrete: cylinders, origin, year, name
```

#### Bar Plot

```python
# Discrete data --> bar plot
# 2D --> xaxis , yaxis
```

```python
df1[df1['origin'] == 1].shape
```

```python
output =  df1['origin'].value_counts()
output
```

```python
uniq = output.index
uniq
```

```python
freq = output.values
freq
```

```python
plt.bar(uniq, freq)
```

```python
uniq
```

```python
plt.bar(uniq, freq, color = ['black','blue','yellow']) # colors will be assigned based on values in uniq [1,3,2]
plt.xlabel('Unique')
plt.ylabel('Freq')
plt.title('Bar Plot for origin')

plt.show() # to hide unwanted information
```

#### Pie chart

```python
# pie chart should be created for discrete data

plt.pie(freq, labels = uniq)
# plt.pie(freq)
plt.show()
```
