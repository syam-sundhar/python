#### Series

```python
# use list / tuple to create a series
# 1D data structure 
# it cant handle mixed data types
# Series(Values, index, dtype)
```

```python
sr1 = pd.Series([10,20,30,40,50]) # default index will be created by 0,1,2,... 
sr1
```

```python
sr1.index
```

```python
type(sr1) # will return data structure of the word sr1 { list/ tuple/int/array...}
```

```python
# numbers : int/float ; strings: object
```

```python
sr1[3]
```

```python
sr2 = pd.Series([10,20,30,40,50], index = ['a','b','c','d','e'], dtype = float)
sr2
```

```python
print(sr2['c'])
```

```python
sr2.index
```

```python
sr2.values
```

```python
sr2.dtype
```

#### DataFrame

```python
# 2D data structure 
# an easy way to create data frame is to combine multiple columns 
# each column of data can be list / tuple/ 1D Array/ Series
```

```python
dict1 = {
    'col1': [10,20,30],
    'col2': pd.Series(['a','b','c'])
}
df1 = pd.DataFrame(dict1)
df1
```

```python
output = df1.dtypes # it returns data types of each column
output
```

```python
type(output)
```

```python
output.index
```

```python
output.values
```

```python
output.dtype
```
