# **Label Encoding**

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
for i in df.columns:
  if df[i].dtypes=='object':
    df[i]=le.fit_transform(df[i])
```

```python
df.info()
```

# **Model Building**

```python
x=df.drop("Income",axis=1)
y=df['Income']
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from imblearn.over_sampling import SMOTE
```

```python
sm=SMOTE(random_state=42)
x_train_bal,y_train_bal=sm.fit_resample(x_train,y_train)
```

# **1. Logistic Regression**

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
```

```python
lr.fit(x_train_bal,y_train_bal)
```

```python
lr_pred=lr.predict(x_test)
```

```python
from sklearn.metrics import *
```

```python
lr_acc=accuracy_score(y_test,lr_pred)
```

```python
lr_acc
```

# **2. Decision Tree**

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt=DecisionTreeClassifier(max_depth=8,random_state=12)
```

```python
dt.fit(x_train_bal,y_train_bal)
```

```python
dt_pred=dt.predict(x_test)
```

```python
dt_acc=accuracy_score(y_test,dt_pred)
```

```python
dt_acc
```

# **3. Random Forest**

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf=RandomForestClassifier()
```

```python
rf.fit(x_train_bal,y_train_bal)
```

```python
rf_pred=rf.predict(x_test)
```

```python
rf_acc=accuracy_score(y_test,rf_pred)
```

```python
rf_acc
```

# **Accuracy Comaprision**

```python
model_names=['Logistic Regression','Decsion Tree','Random Forest']
accuracy_scores=[lr_acc,dt_acc,rf_acc]

plt.figure(figsize=(10,6))
plt.bar(model_names,accuracy_scores)
plt.title('Model Accuracy Comparision')
plt.ylabel('Accuracy Scores')
plt.show()
```

```python
#  Random Forest is giving us the better accuracy
#  So random Forest is the winner.
```
