```python
import numpy as np
import pandas as pd
import matplotlib as plt
import seaborn as sns
```

```python
df=pd.read_csv('csgo_round_snapshots (1).csv')
df
```

```python
df.isnull().sum().sum()
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.duplicated().sum()
```

```python
df.shape
```

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler
le=LabelEncoder()
se=StandardScaler()
```

```python
df['map']=le.fit_transform(df['map'])
```

```python
x=df.drop('round_winner',axis=1)
y=df['round_winner']
```

```python
y=le.fit_transform(y)
```

```python
df.info()
```

```python
x=se.fit_transform(x)
```

```python
x=pd.DataFrame(x)
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_text=train_test_split(x,y,test_size=0.20,random_state=32)
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
lda=LinearDiscriminantAnalysis(n_components=1)
```

```python
x_train_lda=lda.fit_transform(x_train,y_train)
x_test=lda.transform(x_text)
```

```python
x_test.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
re=RandomForestClassifier(n_estimators=50,random_state=23)
```

```python
re.fit(x_train_lda,y_train)
```

```python
y_prid=re.predict(x_test)
```

```python
from sklearn.metrics import *
```

```python
accuracy_score(y_prid,y_text)*100
```

```python
n_estimators=[50,100,150,200,250,300,350,400,450,500,550,600,650]
for i in n_estimators:
    temp_model=RandomForestClassifier(n_estimators=i,random_state=20)
    temp_model.fit(x_train_lda,y_train)
    temp_prid=temp_model.predict(x_test)
    print(f"for n estimator {i} the accuracy score is: {accuracy_score(temp_prid,y_text)}")
```
